```python
Q1 = data['sepal_width'].quantile(0.25)
print(Q1)
Q3 = data['sepal_width'].quantile(0.75)
print(Q3)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
min_=Q1-1.5*IQR
max_=Q3+1.5*IQR
print(min_)
print(max_)
```

```python
print(data.loc[(data['sepal_width']<min_)])
print(data.loc[data['sepal_width']>max_])
```

```python
data
```

```python
# Checking the percentage of outlier in the dataset .
# If it is > 5% then we should not handle it
# If it is < 5% then only we should deal with it.

l = len(data.loc[data['sepal_width']>max_])+len(data.loc[data['sepal_width']<min_])
```

```python
l/len(data)*100
```

```python
# 2.6 means the outliers should be handed.
```

```python
# So now we will impute outliers using median
```

```python
data.loc[data['sepal_width']>max_,'sepal_width'] = data['sepal_width'].median()
```

```python
data.loc[data['sepal_width']>max_]
```

```python
sns.boxplot(x='sepal_width',data=data)              # So now there are no outliers on the right side of the graph which means 
                                                    # We've imputed them in the range using outliers.
plt.plot                                            # Imputing means replacing outliers with mean value of given column.
```

```python
data.loc[data['sepal_width']<min_]
```

```python
data['sepal_width'].mean()
```

```python
data.loc[data['sepal_width']<min_,'sepal_width']=data['sepal_width'].mean()
```
